####<font color=GREEN size=4><b>🔗 Gdrive Downloader</b></font>

```python
# @markdown <font color=GREEN size=5><b>🔧 Gdrive Downloader</b></font>
# @markdown <p>⬤ Change ***mode*** to \"file\" for link from single file link and \"folder\" for link from folder link. Fill ***id_link*** with only the id from url, not full url.</p>
# @markdown <center><img src="https://raw.githubusercontent.com/SudoR2spr/GoFile-Downloader/main/Op%20Source/logo/gdrive.png" alt="GDrive logo" height="50"></center>
import os
# @markdown <font color=GREEN size=4><b>🔧 Gdrive MODE</b></font>
mode = "file" # @param ["file", "folder"]
id_link = "" # @param {type:"string"}
output_path = "/content/drive/My Drive/GoFile" # @param {type:"string"}

import os
delete_path = '/content/sample_data'
if os.path.exists(delete_path):
    !rm -rf '/content/sample_data'

try:
    os.makedirs(output_path, exist_ok=True)
except OSError as error:
    pass
os.chdir(output_path)

if mode == "file":
    command = f"gdown --remaining-ok {id_link}"
else:
    command = f"gdown --folder --remaining-ok {id_link}"

!{command}

os.chdir('/content')
```
